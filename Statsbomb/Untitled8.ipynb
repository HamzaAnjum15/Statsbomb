```python
import matplotlib.pyplot as plt
from mplsoccer import Pitch
from statsbombpy import sb

# Step 1: Get competition, match, lineup, and events data
Competitions = sb.competitions()
Euro = Competitions[(Competitions['competition_name'] == 'UEFA Euro') & (Competitions['season_name'] == '2024')]
Matches = sb.matches(competition_id=55, season_id=282)
Spain = Matches[Matches['home_team'] == 'Spain']
Spain_Lineup = sb.lineups(match_id=3943043)["Spain"]
Spain_events = sb.events(match_id=3943043, split=True, flatten_attrs=False)

# Position mapping
position_coordinates = {
    'Goalkeeper': (1, 50),
    'Right Back': (20, 80),
    'Right Center Back': (40, 80),
    'Center Back': (50, 80),
    'Left Center Back': (60, 80),
    'Left Back': (80, 80),
    'Right Wing': (20, 60),
    'Right Midfield': (40, 60),
    'Center Midfield': (50, 60),
    'Left Midfield': (60, 60),
    'Left Wing': (80, 60),
    'Right Forward': (30, 40),
    'Center Forward': (50, 40),
    'Left Forward': (70, 40),
    'Striker': (50, 30)
}

# Step 2: Extract player positions from the lineup data
player_positions = []
for index, player in Spain_Lineup.iterrows():
    positions = player['positions']
    if isinstance(positions, list) and len(positions) > 0:
        position = positions[0]
        pos_name = position['position']
        pos_coords = position_coordinates.get(pos_name, (0, 0))  # Default to (0, 0) if position not found
        player_positions.append({
            'name': player['player_name'],
            'position': pos_name,
            'x': pos_coords[0],
            'y': pos_coords[1]
        })

# Step 3: Plot the formation using mplsoccer
pitch = Pitch(pitch_type='statsbomb', line_color='black')
fig, ax = pitch.draw()

for player in player_positions:
    pitch.scatter(player['x'], player['y'], ax=ax, s=100, color='blue', edgecolors='black', zorder=3)
    ax.text(player['x'], player['y'] + 2, player['name'], ha='center', va='center', fontsize=10, color='black')

plt.title("Spain Formation")
plt.show()
```

```python
import matplotlib.pyplot as plt
from mplsoccer import Pitch
from statsbombpy import sb

# Step 1: Get competition, match, lineup, and events data
Competitions = sb.competitions()
Euro = Competitions[(Competitions['competition_name'] == 'UEFA Euro') & (Competitions['season_name'] == '2024')]
Matches = sb.matches(competition_id=55, season_id=282)
Spain = Matches[Matches['home_team'] == 'Spain']
Spain_Lineup = sb.lineups(match_id=3943043)["Spain"]
Spain_events = sb.events(match_id=3943043, split=True, flatten_attrs=False)

# Position mapping
position_coordinates = {
    'Goalkeeper': (1, 50),
    'Right Back': (20, 80),
    'Right Center Back': (40, 80),
    'Center Back': (50, 80),
    'Left Center Back': (60, 80),
    'Left Back': (80, 80),
    'Right Wing': (20, 60),
    'Right Midfield': (40, 60),
    'Center Midfield': (50, 60),
    'Left Midfield': (60, 60),
    'Left Wing': (80, 60),
    'Right Forward': (30, 40),
    'Center Forward': (50, 40),
    'Left Forward': (70, 40),
    'Striker': (50, 30)
}

# Step 2: Extract player positions from the lineup data
player_positions = []
for index, player in Spain_Lineup.iterrows():
    positions = player['positions']
    if isinstance(positions, list) and len(positions) > 0:
        position = positions[0]
        pos_name = position['position']
        pos_coords = position_coordinates.get(pos_name, (0, 0))  # Default to (0, 0) if position not found
        player_positions.append({
            'name': player['player_name'],
            'position': pos_name,
            'x': pos_coords[0],
            'y': pos_coords[1]
        })

# Step 3: Plot the formation using mplsoccer
pitch = Pitch(pitch_type='statsbomb', pitch_color='grass', line_color='white',stripe=True)
fig, ax = pitch.draw(figsize=(10, 7))

for player in player_positions:
    pitch.scatter(player['x'], player['y'], ax=ax, s=300, color='blue', edgecolors='black', zorder=3)
    ax.text(player['x'], player['y'] + 2, player['name'], ha='center', va='bottom', fontsize=10, color='white')

plt.title("Spain Formation")
plt.show()
```

```python
import matplotlib.pyplot as plt
from mplsoccer import Pitch

# Create a Pitch object with a Wyscout-style layout
pitch = Pitch(pitch_type='wyscout', pitch_color='#22312b', line_color='#c7d5c9',
              linewidth=2, stripe=False)

# Define the formation data (you can adjust this to match your team's formation)
formation_data = [
    {'player': 'GK', 'x': 0, 'y': 45},
    {'player': 'DF1', 'x': 20, 'y': 40},
    {'player': 'DF2', 'x': 30, 'y': 40},
    {'player': 'DF3', 'x': 40, 'y': 40},
    {'player': 'MF1', 'x': 25, 'y': 30},
    {'player': 'MF2', 'x': 35, 'y': 30},
    {'player': 'MF3', 'x': 45, 'y': 30},
    {'player': 'FW1', 'x': 30, 'y': 20},
    {'player': 'FW2', 'x': 40, 'y': 20},
    {'player': 'FW3', 'x': 50, 'y': 20}
]

# Plot the formation
fig, ax = pitch.draw()
for player in formation_data:
    ax.text(player['x'], player['y'], player['player'], ha='center', va='center', size=12, color='white')
    ax.scatter(player['x'], player['y'], s=200, color='white', zorder=10)

# Customize the plot
ax.set_title('Spain Formation', fontsize=18)
ax.set_xlabel('')
ax.set_ylabel('')
ax.set_xlim(0, 80)
ax.set_ylim(0, 80)

# Show the plot
plt.show()
```
